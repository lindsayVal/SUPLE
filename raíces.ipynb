```python
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625    
```

# GR√ÅFICA

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-20, 20, 300) 
y = f(x)

plt.axhline(0)
plt.plot(x, y)
plt.xlim(-20, 20)
plt.ylim(-10, 10)
plt.show


```
